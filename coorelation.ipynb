```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('visual-taxonomy/train.csv')
df2 = df[df['Category'] == 'Sarees']
df2.head()
```

```python
df2.drop(['Category'], axis=1, inplace=True)
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in range(1, 12):
    df2.iloc[:, i] = le.fit_transform(df2.iloc[:, i])
```

```python
df2.head()
```

```python
df2.isnull()
```

```python
df2.fillna(-1, inplace=True)
```
